```python
# Better more complex architecture of RNN

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense, Dropout
from tensorflow.keras.optimizers import RMSprop

# Define model
model_rnn2 = Sequential([
    SimpleRNN(128, return_sequences=True, input_shape=(TIME_STEPS, FEATURES)),  # Increased units
    Dropout(0.3),  # Regularization
    SimpleRNN(64, return_sequences=True),
    Dropout(0.3),
    SimpleRNN(32),  # Final RNN layer
    Dense(13, activation='softmax')  # Output layer
])

# Compile model with RMSprop and gradient clipping
optimizer = RMSprop(learning_rate=0.001, clipnorm=1.0)
model_rnn2.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])

# Train model
history = model_rnn2.fit(X_train_seq, y_train_seq, epochs=20, batch_size=64, validation_split=0.2)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Load the saved training history
history_data = history.history

# Plot Accuracy
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(history_data['accuracy'], label='Train Accuracy')
plt.plot(history_data['val_accuracy'], label='Validation Accuracy')
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.title("Model Accuracy Over Epochs")

# Plot Loss
plt.subplot(1, 2, 2)
plt.plot(history_data['loss'], label='Train Loss')
plt.plot(history_data['val_loss'], label='Validation Loss')
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.title("Model Loss Over Epochs")

plt.show()
```

```python
from tensorflow.keras.regularizers import l2
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

# Define model
model_rnn4 = Sequential([
    SimpleRNN(256, return_sequences=True, kernel_regularizer=l2(0.0005), input_shape=(TIME_STEPS, FEATURES)),
    Dropout(0.4),
    SimpleRNN(128, return_sequences=True, kernel_regularizer=l2(0.0005)),
    Dropout(0.3),
    SimpleRNN(64),
    Dense(13, activation='softmax')
])

# Optimizer with smaller learning rate
optimizer = RMSprop(learning_rate=0.0005, clipnorm=1.0)

model_rnn4.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])

# Callbacks
callbacks = [
    EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, min_lr=1e-5)
]

# Train model
history = model_rnn4.fit(X_train_seq, y_train_seq, epochs=50, batch_size=32, validation_split=0.2, callbacks=callbacks)
```

```python


# Load the saved training history
history_data = history.history

# Plot Accuracy
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(history_data['accuracy'], label='Train Accuracy')
plt.plot(history_data['val_accuracy'], label='Validation Accuracy')
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.title("Model Accuracy Over Epochs")

# Plot Loss
plt.subplot(1, 2, 2)
plt.plot(history_data['loss'], label='Train Loss')
plt.plot(history_data['val_loss'], label='Validation Loss')
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.title("Model Loss Over Epochs")

plt.show()
```

```python
from sklearn.metrics import accuracy_score, f1_score
import time

# Function to evaluate a model
def evaluate_model(model, X_test, y_test):
    start_time = time.time()
    predictions = model.predict(X_test)
    end_time = time.time()

    # Convert probabilities to class labels
    y_pred = predictions.argmax(axis=1)
    y_true = y_test.argmax(axis=1)

    # Compute metrics
    accuracy = accuracy_score(y_true, y_pred) * 100
    f1 = f1_score(y_true, y_pred, average='weighted') * 100
    inference_time = (end_time - start_time) / len(X_test) * 1000  # ms/sample

    print(f"Accuracy: {accuracy:.2f}%")
    print(f"F1-score: {f1:.2f}%")
    print(f"Inference Time: {inference_time:.3f} ms/sample")

    return accuracy, f1, inference_time

# Evaluate full models
print("Evaluating Full RNN Model:")
evaluate_model(model_rnn4, X_test_seq, y_test_seq)
```
